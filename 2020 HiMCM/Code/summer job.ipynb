```python
# 距离函数
def skill_distance(skill_personal,skill_work,weight=1):# 技能产生的距离
    if int(skill_personal) >= skill_work:# 达到要求
        return 0
    else:
        return weight*abs(int(skill_personal) - skill_work)
    
def Distance_distance(distance_para,weight=1):#distance产生的距离
    return 0

def normal_distance(work,personal,normal=True,weight=1):#常规项产生的距离
    if normal==True:
        if work > personal:# 未达到工作要求
            return abs(weight)*abs(work-personal)
        else:
            return 0
    elif normal==False:
        if work < personal:# 超过工作限定
            return abs(weight)*abs(work-personal)
        else:
            return 0
```

```python
# 排序与汇报
def sort_order(num,reverse=True): # 将得分排序
    num_sort=sorted(num,reverse=reverse)
    order=[0]*len(num)
    before=num_sort[0]
    k=1
    for i in range(len(num_sort)):
        for j in range(len(num)):
            if i>0 and num_sort[i] == before:
                continue
            if num[j] == num_sort[i]:
                order[j]=k
                k=k+1
        before=num_sort[i]
    order_r=[0]*len(num)
    for i in range(len(num)):
        for j in range(len(num)):
            if i+1==order[j]:
                order_r[i]=j+1
                break
    return order,order_r

def report(num,distance_all,works_name,skill,works,n=3,reverse=False):# 对工作进行排序推荐
    distance=distance_all[num]
    skill=skill.iloc[num]
    print("推荐工作排序:\n\t%-20s"%str("work name")+'\tdistance'+'\tskill level(person/work)')
    order,order_r=sort_order(num=distance,reverse=reverse)
    if n>len(works_name):
        n=len(works_name)
    for i in range(n):
        print('%d:'%(i+1)+'\t%-20s'%works_name[order_r[i]-1]+'\t%-10.2f'%distance[order_r[i]-1],end='')
        print('\t%d'%int(skill[works_name[order_r[i]-1]])+'/%d'%int(works[works_name[order_r[i]-1]][5]))
```

```python
# 读取参数
works=pd.read_csv(r'para_work.csv',header=0)      # 读取工作要求参数
works_name=works.columns[2:]  # 工作名
features=works.iloc[:,0]      # 特征名、factors
work_weight=works.iloc[:,1]        # 特征的权重

para=pd.read_csv(r'para_personal.csv')                      # 读取人物的参数
person_name = {(i-65):chr(i) for i in range(65, 65+10)}  # 人物名称
name_list = [chr(i) for i in range(65, 65+10)]             # 人物名称 A-J

skill=pd.read_csv(r'skill_personal.csv',header=0) # 读取假人技能参数
distance_weight=1                      # 距离的权重
skill_para=skill[0::2]                 # 个人信息
skill_weight=skill[1::2].iloc[:,1:]*5 # 不同人物工作权重(所有人物暂相同)
skill_para.index=range(10)             # 设置行名
skill_weight.index=range(10)           # 设置行名

# 计算所有距离
distance_all=[[]]*len(person_name) # 计算人物对所有工作的距离
for m in range(len(person_name)):    
    para_person=para[person_name[m]]
    distance_person=[0]*len(works_name)
    k=0 
    for i in works_name:
        total=0
        for j in range(len(features)):
            if features[j]=='Distance':
                distance=Distance_distance(distance_para=para_person[j],weight=skill_weight[i][j])
            elif features[j]=='Skill Type':
                distance=skill_distance(skill_personal=skill_para[i][m],skill_work=works[i][j],weight=skill_weight[i][j])
            elif work_weight[j]>0:  # 工作要求至少满足该特征
                distance=normal_distance(works[i][j],para_person[j],normal=True,weight=work_weight[j])
            elif work_weight[j]<0:# 工作限制最多为该特征
                distance=normal_distance(works[i][j],para_person[j],normal=False,weight=work_weight[j])
            total=total+distance
        distance_person[k]=total
        k=k+1
    distance_all[m]=distance_person

# 汇报个人工作推荐
num=9 # 设置为第一个人
report(num,distance_all,works_name,skill_para,works,n=303)
```
